### Now need to show group difference

```python
group_data = my_col.transpose()
group_data
```

```python
newIndex = []
stringToReplace = "_01_alpha-lh"

for aGrouper in group_data.index:
    tmp = aGrouper.replace("_01_alpha-lh","")
    newIndex.append(tmp)
    
print(newIndex)
group_data.index = newIndex
```

```python
group_data.index
```

```python
group_data['Group'] = pd.NA
group_data.loc[(group_data.index.astype(int) > 100) & (group_data.index.astype(int) <= 200), 'Group'] = 'MDD'
group_data.loc[group_data.index.astype(int) > 200, 'Group'] = 'HV'
group_data
```

```python
groupmean = group_data.groupby('Group').mean().transpose()
groupstd = group_data.groupby('Group').std().transpose()
```

```python
step = 15

meanMHV = groupmean["HV"] - groupstd["HV"]
meanPHV = groupmean["HV"] + groupstd["HV"]

meanMMDD = groupmean["MDD"] - groupstd["MDD"]
meanPMDD = groupmean["MDD"] + groupstd["MDD"]

plt.figure(figsize=(16, 6))
plt.plot(groupmean.index[::step], groupmean["HV"][::step], label="HV", color="blue")
plt.plot(groupmean.index[::step], groupmean["MDD"][::step], label="MDD", color="green")

x = np.arange(len(groupmean))

plt.fill_between(x[::step], 
                 meanMHV[::step],
                 meanPHV[::step], 
                 color='green', alpha=0.2, label='HV St.Dev')

plt.fill_between(x[::step], 
                 meanMMDD[::step],
                 meanPMDD[::step], 
                 color='blue', alpha=0.2, label='MDD St.Dev')

plt.title('Average Time Series with Standard Deviation')
plt.xlabel('Time')
plt.ylabel('Source Values')
plt.legend()
plt.show()
```

```python
groupCorr = my_col
```

```python
newIndex = []
stringToReplace = "_01_alpha-lh"

for aGrouper in my_col.columns:
    tmp = aGrouper.replace("_01_alpha-lh","")
    newIndex.append(tmp)
    
print(newIndex)
groupCorr.columns = newIndex
```

```python
columns = groupCorr.columns.astype(str)
MDD_columns = [col for col in columns if 100 <= int(col) <= 199]
MDDcorr = groupCorr[MDD_columns]

HV_columns = [col for col in columns if 200 <= int(col) <= 299]
HVcorr = groupCorr[HV_columns]

print(MDD_columns)
print(HV_columns)
```

```python
#print(HVcorr)
#print(MDDcorr)
```

```python
roll_corrHV = HVcorr.rolling(window=5).corr()
roll_corrHV.index.names = ['A','B']

roll_corrMDD = MDDcorr.rolling(window=5).corr()
roll_corrMDD.index.names = ['A','B']

roll_corr_aveMDD = roll_corrMDD.groupby(level='A').mean()
roll_corr_stdMDD = roll_corrMDD.groupby(level='A').std()

roll_corr_aveHV = roll_corrHV.groupby(level='A').mean()
roll_corr_stdHV = roll_corrHV.groupby(level='A').std()

#print(roll_corr_aveMDD)
#print(roll_corr_aveHV)
```

```python
ra_MDD = roll_corr_aveMDD.mean(axis=1)
rs_MDD = roll_corr_stdMDD.mean(axis=1)

ra_HV = roll_corr_aveHV.mean(axis=1) 
rs_HV = roll_corr_stdHV.mean(axis=1)
```

```python
step = 15

plt.figure(figsize=(16, 6))
plt.plot(ra_MDD[::step], label="mean_MDD", color="blue")
plt.plot(ra_HV[::step], label="mean_HV", color="green")

plt.fill_between(ra_HV.index[::step], 
                 ra_HV[::step] - rs_HV[::step], 
                 ra_HV[::step] + rs_HV[::step], 
                 color='green', alpha=0.2, label='HV STD')

plt.fill_between(ra_MDD.index[::step], 
                 ra_MDD[::step] - rs_MDD[::step], 
                 ra_MDD[::step] + rs_MDD[::step], 
                 color='blue', alpha=0.2, label='MDD STD')

plt.title('Average Time Correlation with Standard Deviation')
plt.xlabel('Time')
plt.ylabel('Average Time Series Correlation')
plt.legend()
plt.show()
```
